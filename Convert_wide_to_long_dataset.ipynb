```python
import pandas_ods_reader
from pandas_ods_reader import read_ods
import pandas as pd
```

```python
# define paths:
pathtoData = '/content/drive/Shareddrives/Eye_Tracking_NN_SD/Image_processing/Publishable_material/Datasets_regressions/Wide_format/'
pathtoSave = r'/content/drive/Shareddrives/Eye_Tracking_NN_SD/Image_processing/Publishable_material/Datasets_regressions/Long_format/'
```

```python
datasets = ['Dataset_regression_full', 'Dataset_regression_80p', 'Dataset_regression_50p', 'Dataset_regression_30p', 'Dataset_regression_15p', 'Dataset_regression_15s', 'Dataset_regression_10s', 'Dataset_regression_5s', 'Dataset_regression_2s' ]
```

```python
# Convert datasets form wide to Long format:
for i in range(len(datasets)):
  df = read_ods(pathtoData + datasets[i] + '.ods' , 1)
  # convert caterogy variables from float to interger (e.g. from 1.0 to 1)
  df.iloc[:,10:13] = df.iloc[:,10:13].astype(int)
  levels = df.iloc[:, 10:13].idxmax(axis=1)
  df['Levels'] = levels
  df.to_excel(pathtoSave + datasets[i] + '.xlsx', index = False)
```
