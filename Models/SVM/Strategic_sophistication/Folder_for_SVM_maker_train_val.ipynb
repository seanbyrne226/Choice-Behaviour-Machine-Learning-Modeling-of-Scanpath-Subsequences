#2. Create train and validation images folders

```python
RootDir1 = r'/content/drive/Shareddrives/Eye_Tracking_NN_SD/Image_processing/Publishable_material/Images/Train'
RootDir3 = r'/content/drive/Shareddrives/Eye_Tracking_NN_SD/Image_processing/Publishable_material/Images/Validation'
TargetFolder1 = r'/content/drive/Shareddrives/Eye_Tracking_NN_SD/Image_processing/Publishable_material/Images/Train_SVM_Strategic_sophistication'
TargetFolder3 = r'/content/drive/Shareddrives/Eye_Tracking_NN_SD/Image_processing/Publishable_material/Images/Val_SVM_Strategic_sophistication'
for root, dirs, files in os.walk((os.path.normpath(RootDir1)), topdown=False):
        for name in files:
            if name.endswith('.jpg'):
                print ("Found")
                SourceFolder = os.path.join(root,name)
                shutil.copy2(SourceFolder, TargetFolder1) #copies csv to new folder

for root, dirs, files in os.walk((os.path.normpath(RootDir3)), topdown=False):
        for name in files:
            if name.endswith('.jpg'):
                #print ("Found")
                SourceFolder = os.path.join(root,name)
                shutil.copy2(SourceFolder, TargetFolder3) #copies csv to new folder

```
