```python
import pandas as pd
import numpy as np 
import os 

from sklearn.model_selection import train_test_split
```

### Shuffle


```python
"""

This script outputs 3 CSV files created from the original 'codes_no_DS_test_31_5_2021.csv'. 
These 3 outputs correspond to Test, Validation and Train dataset used for the classification.

"""
```

```python
### set parent path 
path = '/content/drive/Shareddrives/Eye_Tracking_NN_SD/Image_processing/Publishable_material/'

### set save folder name  
saveFolder = 'data/'
```

```python
### load 'codes_no_DS_test_31_5_2021.csv' file 
df = pd.read_csv(path + 'codes_no_DS_test_31_5_2021.csv')
indexTest = df[ df['phase']=='phase-1'].index

### make folder to save files in
os.mkdir(path + saveFolder)

### path of folder to save output csv files in data
pathSave = path + saveFolder
```

```python
### Make df for only Phase = 1 
df2 = df.copy()
df2 = df2.iloc[indexTest,:].copy()
df2 = df2.drop(['L2', 'Coop', 'L1', 'Test'], axis=1)
print('Shape of df only phase = 1 : ' + str(df2.shape))
print('Name of columns of DF ' +str(df2.columns))
```

```python
#Get Subject numbers and split them in to Train(70%), Test(20%), Val(10%) 
uniqueValues = df2['sbj'].unique()
train, testVal = train_test_split(uniqueValues, test_size=0.3, train_size=0.7, random_state=28,  shuffle=True)
Val, test = train_test_split(testVal, test_size=0.33, train_size=0.66, random_state=28,  shuffle=True)

print('Total: ' + str(len(uniqueValues)) +'  , '+ str(100*len(uniqueValues)/len(uniqueValues)) +'%')
print('Train: ' + str(len(train)) +'  , '+ str(round( 100*len(train)/len(uniqueValues), 2)) +'%')
print('Val: ' + str(len(Val)) +'  , '+ str(round( 100*len(Val)/len(uniqueValues), 2)) +'%')
print('Test: ' + str(len(test)) +'  , '+ str(round( 100*len(test)/len(uniqueValues), 2)) +'%')
```

```python
# make train CSV

frames= []
cmaps = ['spring', 'summer', 'autumn', 'winter', 'cool', 'Wistia']

for i in train:
  frames.append(df2[ df2['sbj']==i])
  
trainCSV = pd.concat(frames)
trainCSV = trainCSV.reset_index()
trainCSV = trainCSV.drop(['index'], axis=1)

trainCSV.to_csv(pathSave + 'train.csv', index=None)

print('Shape Train 1: ' +str(trainCSV.shape))
print(trainCSV.head())
```

```python
# make val CSV
frames= []

for i in Val:

  frames.append(df2[ df2['sbj']==i])
  
valCSV = pd.concat(frames)
valCSV = valCSV.reset_index()
valCSV = valCSV.drop(['index'], axis=1)

valCSV.to_csv(pathSave + 'validation.csv', index=None)

print('Shape val: ' +str(valCSV.shape))
print(valCSV.head())
```

```python
# make test CSV
frames= []

for i in test:

  frames.append(df2[ df2['sbj']==i])
  
testCSV = pd.concat(frames)
testCSV = testCSV.reset_index()
testCSV = testCSV.drop(['index'], axis=1)

testCSV.to_csv(pathSave + 'test.csv', index=None)

print('Shape test: ' +str(testCSV.shape))
print(testCSV.head())
```
